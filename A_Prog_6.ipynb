```python
#1
def enter_values():
    """
    enter values 2 strings
    """
    s = input("enter a string")
    t = input("enter a string with additional one character")
    return s,t
def find_the_diff(s,t):
    """
    String t is generated by randomly
shuffling string s and then adding one more letter at a random position.
Return the letter that was added to t.
    """
    for i in range(len(t)):
        if (t[i] not in s) | (t.count(t[i]) > s.count(t[i])): 
            return t[i]
    return " "
s,t=enter_values()
print(find_the_diff(s,t))
```

```python
#2
def count_datatypes(l):
    """
    count [int, str, bool, list, tuple, dictionary] in l
    """
    lis,ii,s,b,ll,t,d=[],0,0,0,0,0,0
    for i in l:
        if type(i) == int:
            ii=ii+1
        if type(i) == str:
            s=s+1
        if type(i) == bool:
            b=b+1
        if type(i) == list:
            ll=ll+1 
        if type(i) == tuple:
            t=t+1
        if type(i) == dict:
            d=d+1
    lis =[ii,s,b,ll,t,d]
    return lis
l = eval(input("enter any random value"))
print(count_datatypes(l))
```

```python
#3
def enter_values():
    """
    enter values
    """
    n = int(input("enter the fibonnaci length"))
    l=[]
    l.append(input("enter 1st character"))
    l.append(input("enter 2nd character"))
    return n,l
def fib_str(n,l):
    """
    It works with
any two characters of the English alphabet (as opposed to the numbers 0 and
1 in the Fibonacci series) as the initial items and concatenates them together
as it progresses in a similar fashion as the Fibonacci series.
    """
    i = n-2
    a,b= 0,1
    while i>0:
        l.append(l[a]+l[b])
        c=a+b
        a,b=b,c
        i=i-1
    return l
n,l = enter_values()
print(fib_str(n,l))
```
